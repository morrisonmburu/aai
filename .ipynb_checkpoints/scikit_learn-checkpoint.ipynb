#scikit_learn is an open source python library, implements a range of machine learning, preprocessing, cross_validation and visualization algorithms using unified interface

```python
# Encoding categorical data
from sklearn import datasets, neighbors
from sklearn.metrics import accuracy_score
# Encoding the Independent Variable
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
#object fpor encoding independent variables
iris = datasets.load_iris()
X, y = iris.data[:, :4], iris.target

# labelencoder_X_1 = LabelEncoder() 
#identifying the indexes of the categories
# X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1]) 

# #object fpor encoding independent variables
# labelencoder_X_2 = LabelEncoder() 
# #identifying the indexes of the categories
# X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2]) 

# #create the dummy variables
# onehotencoder = OneHotEncoder(categorical_features = [1])
# X = onehotencoder.fit_transform(X).toarray()
# #remove one dummy variable
# X = X[:, 1:]

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 33)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#accuracy score
knn = neighbors.KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```
